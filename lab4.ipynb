```python
print("Задание 1. Вычислить квадраты чисел 1–10 с map.")
nums = list(range(1,11))
squares = list(map(lambda x: x**2, nums))
print("Было: ", nums)
print("Стало: ", squares)
```

```python
from functools import reduce
print("Задание 2. Сумма чисел с редисом. ")
nums = list(range(1,11))
sum_all = reduce(lambda x, y: x+y, nums)
print("Было: ", nums)
print("Стало: ", sum_all)
```

```python
print("Задание 3. Создать генератор Фибоначчи. ")
def fibonacci_generator():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

fibonacci_gen = fibonacci_generator()
fibonacci_numbers = [next(fibonacci_gen) for _ in range(10)]
print("Генератор Фибоначчи: ", fibonacci_numbers)
```

```python
import time
from functools import reduce
print("Задание 4. Составить композицию функций (compose).")
def book_catalog_generator():
    books = [
        {'Название': '1984', 'жанр': 'Антиутопия', 'рейтинг': 4.3, 'длина': 328},
        {'Название': 'Игра престолов', 'жанр': 'Фэнтези', 'рейтинг': 4.2, 'длина': 694},
        {'Название': 'Автостопом по галактике', 'жанр': 'Научная фантастика', 'рейтинг': 4.9, 'длина': 640},
        {'Название': 'Уцелевший', 'жанр': 'Современная проза', 'рейтинг': 4.9, 'длина': 380},
        {'Название': 'Бесконечная шутка', 'жанр': 'Современная проза', 'рейтинг': 4.7, 'длина': 1280},
        {'Название': 'Проклятые', 'жанр': 'Современная проза', 'рейтинг': 4.8, 'длина': 290},
    ]
    for book in books:
        yield book

def timer(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        print("Начало отбора книг...")
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"Решение принято за: {end_time - start_time:.15f} сек")
        return result
    return wrapper
@timer
def find_best_book(catalog_generator):
    eligible_books = filter(
        lambda book: book['жанр'] == 'Современная проза' and book['рейтинг'] > 4.5,
        catalog_generator
    )
    print("1. Фильтрация по направлению ('Современная проза') и рейтингу (> 4.5).")
    scored_books = map(
        lambda book: {**book, 'score': book['рейтинг'] * book['длина']},
        eligible_books
    )
    print("2. Рассчитан коэффициент привлекательности (score).")
    try:
        best_book = reduce(
            lambda winner, current: winner if winner['score'] > current['score'] else current,
            scored_books
        )
        return best_book
    except TypeError:
        return "Не найдено книг, соответствующих заданным критериям."
best_book_data = find_best_book(book_catalog_generator())

print("\nРекомендованная книга: ")
if isinstance(best_book_data, dict):
    print(f"   Название: {best_book_data['Название']}")
    print(f"   Жанр: {best_book_data['жанр']}")
    print(f"   Рейтинг: {best_book_data['рейтинг']}")
    print(f"   Страниц: {best_book_data['длина']}")
    print(f"   Проверяю итоговый коэф привлекательности: {best_book_data['score']:.2f}")
else:
    print(best_book_data)
```

```python
print("Задание 5. Реализовать фильтрацию данных из файла через генератор.")
from functools import reduce
def record_parser(filename):
    try:
        with open(filename, 'r', encoding='utf-8') as f:
            while True:
                line1 = f.readline()
                if not line1:
                    break
                line2 = f.readline()
                if not line2:
                    break

                key1, value1 = line1.strip().split(': ', 1)
                key2, value2 = line2.strip().split(': ', 1)
                yield {key1: value1, key2: value2}
    except FileNotFoundError:
        print(f"Ошибка: Файл '{filename}' не найден.")
    except Exception as e:
        print(f"Произошла ошибка при парсинге: {e}")
records_gen = record_parser("records.txt")
TARGET_DATE = "18.07.2025"

print(f"\n Поиск записи за дату: {TARGET_DATE}...\n")
filtered_records = filter(
    lambda record: record.get('Запись') == TARGET_DATE,
    records_gen
)
filtered_groups = list(map(
    lambda record: record['Группа'],
    filtered_records
))

if filtered_groups:
    print("Найденные группы:")
    for group in filtered_groups:
        print(f"-> {group}")
else:
    print("Записей за эту дату не найдено.")
```
