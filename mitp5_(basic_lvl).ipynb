```python
# -*- coding: cp1251 -*-
import tkinter as tk

message_var = None
def update_message():
    global message_var

    message_var.set("Привет, user!")

    hello_button.config(text="Уже здоровались!")

root = tk.Tk()
root.title("Первое задание")

message_var = tk.StringVar()
message_var.set("Жма на кнопку, чтобы поздороваться")

message_label = tk.Label(
    root,
    textvariable=message_var,
    font=('Courier New', 14),
    fg="olive"
)

message_label.pack(pady=(20, 10))

hello_button = tk.Button(
    root,
    text="Привет-привет",
    command=update_message
)
hello_button.pack(padx=50, pady=20)
root.mainloop()
```

```python
# -*- coding: cp1251 -*-
import tkinter as tk

class Colors:
    PINK_BG = "#FFC0CB"
    HOT_PINK = "#FF69B4"
    SOFT_PINK = "#FADADD"
    TEXT_COLOR = "black"
    NICE_FONT = 'Times New Roman'
    FONT_SIZE_ENTRY = 22
    FONT_SIZE_BUTTON = 16

class CalculatorApp:
    def __init__(self, master):

        self.master = master
        master.title("Калькулятор :3")
        master.resizable(False, False)
        master.configure(bg=Colors.PINK_BG)

        self.current_expression = ""
        self.input_text = tk.StringVar()

        self._create_input_field()
        self._create_buttons()
        self._create_clear_button()

    def _create_input_field(self):
        self.input_field = tk.Entry(
            self.master,
            textvariable=self.input_text,
            font=(Colors.NICE_FONT, Colors.FONT_SIZE_ENTRY, 'bold'),
            bd=5,
            relief=tk.FLAT,
            justify='right',
            bg="white",
            fg=Colors.TEXT_COLOR
        )

        self.input_field.grid(
            row=0, column=0, columnspan=4, sticky='nsew', ipady=15, padx=5, pady=5
        )

    def _create_buttons(self):
        buttons_layout = [
            '7', '8', '9', '/',
            '4', '5', '6', '*',
            '1', '2', '3', '-',
            '0', '.', '=', '+'
        ]

        row_val = 1
        col_val = 0

        for button_text in buttons_layout:
            btn_bg = Colors.HOT_PINK
            command_func = None

            if button_text == '=':

                command_func = self.calculate
                btn_bg = Colors.SOFT_PINK
            elif button_text in ('/', '*', '-', '+', '.'):

                command_func = lambda x=button_text: self.button_click(x)
            else:
                command_func = lambda x=button_text: self.button_click(x)

            tk.Button(
                self.master,
                text=button_text,
                padx=20,
                pady=20,
                font=(Colors.NICE_FONT, Colors.FONT_SIZE_BUTTON, 'bold'),
                command=command_func,
                bg=btn_bg,
                fg=Colors.TEXT_COLOR,
                activebackground=Colors.PINK_BG,
                relief=tk.RAISED
            ).grid(row=row_val, column=col_val, sticky='nsew', padx=3, pady=3)

            col_val += 1
            if col_val > 3:
                col_val = 0
                row_val += 1

        self._last_row_val = row_val

    def _create_clear_button(self):
        tk.Button(
            self.master,
            text='C',
            padx=20,
            pady=20,
            font=(Colors.NICE_FONT, Colors.FONT_SIZE_BUTTON, 'bold'),
            command=self.clear,
            bg=Colors.SOFT_PINK,
            fg=Colors.TEXT_COLOR,
            relief=tk.RAISED
        ).grid(row=self._last_row_val, column=0, columnspan=4, sticky='nsew', padx=3, pady=3)

    def button_click(self, item):
        self.current_expression += str(item)
        self.input_text.set(self.current_expression)

    def clear(self):
        self.current_expression = ""
        self.input_text.set("")

    def calculate(self):
        try:
            result = str(eval(self.current_expression))
            self.input_text.set(result)
            self.current_expression = result
        except ZeroDivisionError:
            self.input_text.set("Деление на 0")
            self.current_expression = ""
        except:
            self.input_text.set("Ошибка")
            self.current_expression = ""

if __name__ == "__main__":
    root = tk.Tk()
    app = CalculatorApp(root)
    root.mainloop()
```
